```python
def show_board(pieces):
    player1 = pieces[0]
    player2 = pieces[1]
    for i in range(16, -1, -1):
        for j in range(17):
            if [i, j] in player1:
                print("1", end=" ")
            elif [i, j] in player2:
                print("2", end=" ")
            else:
                print(".", end=" ")
        print()
```

```python
agents = [RandomAgent(), RandomAgent()]
g = Game(agents)
i = 1
while not g.has_ended():
    if i % 100000 == 0:
        a = g.state.to_json()["pieces"]
        show_board(a)
        print(g.state.to_json())
        print(i)

    i += 1
    g = g.play()
print("YOUPI")
a = g.state.to_json()["pieces"]
show_board(a)
```
